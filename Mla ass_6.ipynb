```python
#Name: soham pokharkar
#Roll no.:D3
#practical no.6 :Implement NaÃ¯ve Bayes theorem to classify the English text

from collections import defaultdict
import re

class NaiveBayesClassifier:
    def __init__(self):
        self.classes = set()
        self.word_counts = defaultdict(lambda: defaultdict(int))
        self.class_counts = defaultdict(int)

    def train(self, data):
        total_words = set()
        for text, label in data:
            self.classes.add(label)
            self.class_counts[label] += 1
            words = re.findall(r'\w+', text.lower())
            for word in words:
                self.word_counts[label][word] += 1
                total_words.add(word)


        for label in self.classes:
            for word in total_words:
                self.word_counts[label][word] += 1
            self.class_counts[label] += len(total_words)

    def predict(self, text):
        words = re.findall(r'\w+', text.lower())
        max_prob = float('-inf')
        predicted_class = None
        for label in self.classes:
            log_prob = 0
            for word in words:
                word_count = self.word_counts[label][word]
                word_total = sum(self.word_counts[label].values())
                word_prob = (word_count + 1) / (word_total + len(self.word_counts[label]))
                log_prob += word_prob
            log_prob += self.class_counts[label] / sum(self.class_counts.values())
            if log_prob > max_prob:
                max_prob = log_prob
                predicted_class = label
        return predicted_class


# Example usage:
data = [("I love this movie", "positive"), ("I hate this movie", "negative")]
classifier = NaiveBayesClassifier()
classifier.train(data)
print(classifier.predict("I hate this movie"))
```

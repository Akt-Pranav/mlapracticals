###Practical No 7 Part-I Unsupervised Learning : Implement K-Means Clustering Problem Statment : Catagorize the Coutries into Continentwise Clusters  on the basis of their Latitude and Longitude


```python
import pandas as pd
data=pd.read_csv('/content/countries (1).csv')
data
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Latitude'],data['Longitude'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

```python
x=pd.DataFrame(data[['Latitude','Longitude']])
x
```

```python
from sklearn.cluster import KMeans
km=KMeans(3)
km.fit(x)
```

```python
clusters=km.fit_predict(x)
clusters
```

```python
data_with_output=data.copy()
data_with_output
```

```python
data_with_output['Cluster']=clusters
data_with_output
```

```python
Continents = []

for i in range(len(clusters)):
    if clusters[i] == 0:
        Continents.append("AMERICA")
    elif clusters[i] == 1:
        Continents.append("EUROPE")
    else:
        Continents.append("AUSTRALIA")

data_with_output['Continent']=Continents
data_with_output
```

```python
plt.scatter(data_with_output['Latitude'], data_with_output['Longitude'],c=data_with_output['Cluster'], cmap='rainbow')
```

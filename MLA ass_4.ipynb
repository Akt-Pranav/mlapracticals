Practical No. 4 Classify nine combinations of VAR1 and VAR2 using K-Means Clustering


```python
import pandas as pd
data=pd.read_csv('/content/data1.csv')
data
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['VAR1'],data['VAR2'])
plt.xlabel('VAR1')
plt.ylabel('VAR2')
plt.show()
```

```python
x=pd.DataFrame(data[['VAR1','VAR2']])
x
```

```python
from sklearn.cluster import KMeans
km=KMeans(3)
km.fit(x)
```

```python
clusters=km.fit_predict(x)
clusters
```

```python
data_with_output=data.copy()
data_with_output
```

```python
data_with_output['Cluster']=clusters
data_with_output
```

```python
plt.scatter(data_with_output['VAR1'], data_with_output['VAR2'],c=data_with_output['Cluster'], cmap='rainbow' )
```

```python
data=pd.read_csv('/content/sat_gpa.csv')
data
```

```python
data['SAT']
```

```python
data.describe()
```

```python
y=data['GPA']
x1=data['SAT']

plt.scatter(x1,y)
plt.xlabel('SAT Data')
plt.ylabel('GPA Data')
plt.show()
```

```python
x=sm.add_constant(x1)

results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat=0.0017 * x1 + 0.2750 # Regression Equation
plt.plot(x1,yhat) # Plotting the Regression Line
plt.xlabel('SAT')
plt.ylabel('GPA')
```
